**Membuat tabel menggunakan Python**

```python
import sqlite3

# membuat koneksi SQL ke database SQLite
con = sqlite3.connect("pets.db", timeout=10)

cur = con.cursor()

# digunakan untuk membuat tabel pet
for row in cur.execute("CREATE TABLE Pet (PetID int, PetName varchar(25), OwnerName varchar(100))"):
    print(row)
con.commit()

# pastikan untuk menutup koneksi
con.close()
```

**Relation mapping in database**

```python
import sqlite3

# membuat koneksi SQL ke database SQLite
con = sqlite3.connect("pets.db", timeout=10)

cur = con.cursor()

# digunakan untuk membuat tabel PetTreatment
cur.execute("PRAGMA foreign_keys = 1")
sql = """CREATE TABLE PetTreatment (PetID int, TreatDate date, TreatResult varchar(255), CONSTRAINT fk_pet FOREIGN KEY (PetID) REFERENCES Pet(PetID))"""
cur.execute(sql)
con.commit()

# pastikan untuk menutup koneksi
con.close()
```

```python
import sqlite3

# membuat koneksi SQL ke database SQLite
con = sqlite3.connect("pets.db", timeout=10)

cur = con.cursor()

# untuk menampilkan isi dari tabel pet dan pettreatment yang berelasi
for row in cur.execute("SELECT p.PetName, p.OwnerName, t.TreatResult FROM Pet as p join PetTreatment as t ON p.PetID = t.PetID"):
    print(row)
con.commit()

# pastikan untuk menutup koneksi
con.close()
```

```python
import sqlite3

# membuat koneksi SQL ke database SQLite
con = sqlite3.connect("pets.db", timeout=10)

cur = con.cursor()

rows = cur.execute("SELECT p.PetName, p.OwnerName, t.TreatDate, t.TreatResult FROM Pet as p JOIN PetTreatment as t ON p.PetID = t.PetID").fetchall()
for row in rows:
    PetName = row[0]
    OwnerName = row[1]
    TreatDate = row[2]
    TreatResult = row[3]
    print("Nama " + PetName + " milik " + OwnerName + " pada " + TreatDate + " mendapatkan " + TreatResult)

con.commit()

# pastikan untuk menutup koneksi
con.close()
```

**RDBMS dan DATAFRAMES**

```python
import pandas as pd
column = ["PetID", "PetName", "OwnerName", "Age", "Gender", "TreatDate", "TreatResult"]
data = []
```

```python
import sqlite3

# membuat koneksi SQL ke database SQLite
con = sqlite3.connect("pets.db", timeout=10)

cur = con.cursor()

# untuk menampilkan isi dari tabel pet dan pettreatment yang berelasi
rows = cur.execute("SELECT p.PetID, p.PetName, p.OwnerName, p.Age, p.Gender, t.TreatDate, t.TreatResult FROM Pet as p JOIN PetTreatment as t ON p.PetID = t.PetID")
for row in rows:
    data.append(row)
con.commit()

# pastikan untuk menutup koneksi
con.close()
```
